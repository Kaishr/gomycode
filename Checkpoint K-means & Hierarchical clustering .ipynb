```python
import pandas as pd
data=pd.read_csv('C:/Users/kaism/Desktop/csv/CC GENERAL.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(),inplace=True)
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(),inplace=True)
```

```python
data.isnull().sum().sum()
```

```python
data=data.drop("CUST_ID",axis=1)
```

```python
data
```

# Hierarchical clustering

```python
#identify the inherent groupings using AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
clust_labels=model.fit_predict(data)
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
#Agglomerative cluster plot
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
scatter=ax.scatter(data["CREDIT_LIMIT"],data["PURCHASES"],c=agglomerative[0],s=50)
ax.set_title("Customer segmentation")
ax.set_xlabel("CREDIT_LIMIT")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

```python
#Dendrogram Example
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data,method="complete"))
```

# K-means Algorithm

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(data)
#K-means Labels
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
#K-means Centroids
kmeans.predict(data)
print(kmeans.cluster_centers_)
```

```python
#K-best Value
Sum_of_squared_distances=[]
K=range(1,10)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(data)
    Sum_of_squared_distances.append(km.inertia_)
print(Sum_of_squared_distances)
```

```python
#The Elbow method code:
plt.plot(K,Sum_of_squared_distances)
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow method for optimal  k')
plt.show()
```
